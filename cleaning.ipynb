```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Data/air_system_previous_years.csv",dtype=object)
```

```python
df.head()
```

```python
#filtering data
df.drop_duplicates()#there is no duplicates, but I had to verify
df = df.replace("na",np.nan)
```

```python
data_ = df.iloc[:,1:]
class_ = df.iloc[:,0]
data_
```

```python
base = pd.concat([class_,data_.apply(pd.to_numeric)],axis=1)
type(base["ae_000"][1])
```

```python
base
```

```python
#using the function to filter the columns that have many zeros
from localFilter import isRemovable
new_base = pd.DataFrame()
for column in base.iloc[:,1:].columns:
    if not isRemovable(base[column],50,checkFor="zeros"):
        new_base[column] = base[column] 
new_base
```

```python
#using the function to filter the columns that have many zeros
from localFilter import isRemovable
new_new_base = pd.DataFrame()
for column in new_base.iloc[:,1:].columns:
    if not isRemovable(base[column],50,checkFor="nan"):
        new_new_base[column] = new_base[column] 
new_new_base
```

```python
#Dropping rows with NaN values
whole_clean_base = pd.concat([base.iloc[:,0],new_new_base],axis=1).dropna()
clean_base = whole_clean_base.iloc[:,1:]
clean_base
```

```python
#calculating correlations by column
corr_matrix = clean_base.corr().abs()
corr_matrix
```

```python
#creating a mask for the triangular supperior part of the triangular matrix
ones_matrix = np.ones_like(corr_matrix,dtype=bool)
boolMask = ~np.triu(ones_matrix,1)#a mask with ones above the main diagonal)
mask = pd.DataFrame(boolMask, index=corr_matrix.index, columns=corr_matrix.columns)
mask
```

```python
corr_matrix.mask(mask)
```

```python
#calcuating the columns to drop and dropping them
corr_limit = 0.7#higth correlation
'''
Explaining this code:
corr_matrix[column] will give a column of name `column` with its correlation in relation with the oder columns.
corr_matrix[column][mask] will give a column of name `column` truncated with the main diagonal
any(corr_matrix[column][mask]>=corr_limit) will return True if any value of the column if any value is bigger than corr_matrix
'''
columns_list =[column for column in corr_matrix.columns if any(corr_matrix.mask(mask)[column]>=corr_limit)]
columns_list
```

```python
cleaner_base = pd.DataFrame(clean_base[[column for column in clean_base.columns if column not in columns_list]])
cleaner_base = pd.concat([whole_clean_base.iloc[:,0],cleaner_base],axis=1)
cleaner_base
```

```python
#Now that we know the correlated columns we go back and first delete those columns and the we drop the nan values, that way we won around 4000 data points without violating the purpouse of the cleaning
tmp = pd.concat([base.iloc[:,0],new_new_base],axis=1)
other_cleaner_base = pd.DataFrame(tmp[[column for column in tmp.columns if column not in columns_list]]).dropna()
other_cleaner_base
```

```python
#saving into files
cleaner_base.to_csv('Data/cleaned1_air_system_previous_years.csv',sep=',',index_label=False)
other_cleaner_base.to_csv('Data/.csv',sep=',',index_label=False)
```
